```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pickle
import seaborn as sns
```

```python
import matplotlib as mpl
from matplotlib.colors import ListedColormap

N = 256
vals = np.ones((N, 4))
vals[:, 0] = np.linspace(74/256 + (189-74)/10/256, 189/256,  N)
vals[:, 1] = np.linspace(98/256 + (98-65)/10/256, 65/256, N)
vals[:, 2] = np.linspace(189/256 + (65-189)/10/256, 65/256,  N)
vals[:, 3] *= 0.75
newcmp = ListedColormap(vals)
```

```python
resdict_con4_s2prop = pickle.load(open( "results/Resultdicts_4_hetstat_9.0_global_changeprop.p", "rb" ))
resdict_vb4_s2prop = pickle.load(open( "results/Resultdicts_4_hetstat_9.0_composite_changeprop.p", "rb" ))
```

```python
mu_con4_s2prop = {}
key_arr = [0.5]
lss = []

for j in key_arr:
    mu_con4_s2prop[j] = []
    for i in resdict_con4_s2prop[j]:
        mu_con4_s2prop[j].append(i.mean)
    mu_con4_s2prop[j] = np.array(mu_con4_s2prop[j])

fig, axs = plt.subplots(4, sharex=True, sharey=True, figsize=(4,6))
for k in range(4):
    sns.kdeplot(data=mu_con4_s2prop[0.5][:,k],color =newcmp(j*2), ls = '-',label = str(j), bw_adjust = 0.1/2,
               ax=axs[k])
    plt.ylim(0,100)
    plt.xlim(-0.2,1.2)

plt.title('Global Static')
plt.tight_layout()
```

```python
mu_vb4_s2prop = {}
key_arr = [0.5]

for j in key_arr:
    mu_vb4_s2prop[j] = []
    for i in resdict_vb4_s2prop[j]:
        mu_vb4_s2prop[j].append(i.mean)
    mu_vb4_s2prop[j] = np.array(mu_vb4_s2prop[j])
    
fig, axs = plt.subplots(4, sharex=True, sharey=True, figsize=(4,6))
for k in range(4):
    sns.kdeplot(data=mu_vb4_s2prop[0.5][:,k],color =newcmp(j*2), ls = '-',label = str(j), bw_adjust = 0.1/2, 
                ax=axs[k])
    plt.ylim(0,750)
    plt.xlim(-0.2,1.2)

plt.title('Composite Static')
plt.tight_layout()
```

```python
res_h_s2prop = pickle.load(open( "Resultdicts_4_hetstat_9.0_hhmm_changeprop.p", "rb" ))
```

```python
m = {}

for resk in res_h_s2prop.keys():
    m[resk] = []
    
    for ress in res_h_s2prop[resk]:
        m[resk].append(ress.mean)
    m[resk] = np.array(m[resk])

fig, axs = plt.subplots(4, sharex=True, sharey=True, figsize=(4,6))

for k in range(4):
    sns.kdeplot(data=m[0.5][:,k],color =newcmp(0.5*2), ls = '-',label = str(0.5), bw_adjust = 0.1/2, ax=axs[k])
    plt.ylim(0,5000)
    plt.xlim(-0.2,1.2)
    
plt.title('Hierarchical Static')
plt.tight_layout()
```
